```python
import nltk
from spellchecker import SpellChecker
```

```python
spell = SpellChecker(SpellChecker.languages())
```

```python
with open('derprozess.txt', encoding='utf-8') as f:
  text = f.read()
  tokenizer = nltk.RegexpTokenizer('\w+\'*\w*')  
  tokens = tokenizer.tokenize(text)
#   print(text)
  unknown = spell.unknown(tokens)
  print(unknown)
```
